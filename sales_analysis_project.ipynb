# Sales Data Analysis Project
This notebook demonstrates data cleaning, exploratory analysis, and visualization using a sales dataset.

## Steps:
1. Load dataset
2. Clean data
3. Analyze sales trends
4. Visualize insights


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset (make sure CSV is in same folder)
df = pd.read_csv('sales.csv')

df.head()
```

```python
# Basic info
df.info()
```

```python
# Data Cleaning
df = df.drop_duplicates()

# Convert Order Date column to datetime (if exists)
if 'Order Date' in df.columns:
    df['Order Date'] = pd.to_datetime(df['Order Date'])

df.isnull().sum()
```

```python
# Monthly Sales Trend
if 'Order Date' in df.columns and 'Sales' in df.columns:
    df['Month'] = df['Order Date'].dt.month
    monthly_sales = df.groupby('Month')['Sales'].sum()
    monthly_sales.plot(kind='line')
    plt.title('Monthly Sales Trend')
    plt.show()
```

```python
# Top Selling Products
if 'Product Name' in df.columns and 'Sales' in df.columns:
    top_products = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)
    top_products.plot(kind='bar')
    plt.title('Top Products')
    plt.show()
```

```python
# Category Wise Sales
if 'Category' in df.columns and 'Sales' in df.columns:
    category = df.groupby('Category')['Sales'].sum()
    category.plot(kind='pie', autopct='%1.1f%%')
    plt.show()
```
